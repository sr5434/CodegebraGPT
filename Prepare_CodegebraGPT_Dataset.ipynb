```python
from datasets import load_dataset
math = load_dataset("meta-math/MetaMathQA")
math2 = load_dataset("camel-ai/math")
math3 = load_dataset("ArtifactAI/arxiv-math-instruct-50k")
chemistry = load_dataset("camel-ai/chemistry")
bio = load_dataset("camel-ai/biology")
physics = load_dataset("camel-ai/physics")
physics2 = load_dataset("ArtifactAI/arxiv-physics-instruct-tune-30k")
gsm = load_dataset("gsm8k", "main")
mmlu = load_dataset("cais/mmlu", "all")
code = load_dataset("nickrosh/Evol-Instruct-Code-80k-v1")
code2 = load_dataset("glaiveai/glaive-code-assistant-v2")
cs = load_dataset("ArtifactAI/arxiv-cs-ml-instruct-tune-50k")
#code3 = load_dataset("nampdn-ai/tiny-codes")
sci = load_dataset("cnut1648/ScienceQA-LLAVA")
```

```python
def format_mmlu(question, choices):
  alphabet = list("abcdefghijklmnopqrstuvwxyz")
  options = ""
  for i in range(len(choices)):
    options += f"{alphabet[i]}. {choices[i]}\n"
  return f"""{question}\n{options}"""
```

```python
inputs = math["train"]["query"] + mmlu["auxiliary_train"].map(lambda batch: {"query": format_mmlu(batch["question"], batch["choices"])})["query"] + code["train"]["instruction"] + code2["train"]["question"] + gsm["train"]["question"] + math2["train"]["message_1"] + math3["train"]["question"] + physics2["train"]["question"] + cs["train"]["question"] + physics["train"]["message_1"]  + chemistry["train"]["message_1"] + bio["train"]["message_1"]
outputs = math["train"]["response"] + list(map(lambda a: list("abcdefghijklmnopqrstuvwxyz")[a], mmlu["auxiliary_train"]["answer"])) + code["train"]["output"] + code2["train"]["answer"] + gsm["train"]["answer"] + math2["train"]["message_2"] + math3["train"]["answer"] + physics2["train"]["answer"] + cs["train"]["answer"] + physics["train"]["message_2"]  + chemistry["train"]["message_2"] + bio["train"]["message_2"]
```

```python
json_data = []
for i in range(len(inputs)):
  json_data.append({
      "id": i,
      "image": "",
      "conversations": [
          {
              "from": "human",
              "value": inputs[i]
          }, {
              "from": "gpt",
              "value": outputs[i]
          }
      ]
  })
```

```python
for i in range(len(sci["train"])):
  if sci["train"][i]["image"] != None:
    sci["train"][i]["image"].save(f"./images/{i}.png")
```

```python
import os
for i in range(len(sci["train"])):
  if os.path.isfile(f"./images/{i}.png"):
    json_data.append({
        "id": len(inputs) + i,
        "image": f"./images/{i}.png",
        "conversations": sci["train"][i]["conversations"]
    })
  else:
    json_data.append({
        "id": len(inputs) + i,
        "image": "",
        "conversations": sci["train"][i]["conversations"]
    })
```

```python
from random import shuffle
shuffle(json_data)
```

```python
import json

with open("train.json", 'w') as fout:
    print(json.dumps(json_data[:100000], indent=4), file=fout)
```
